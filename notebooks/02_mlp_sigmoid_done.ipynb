## 🏗️ Arsitektur
1 hidden layer berisi **16 neuron**, lalu output 3 neuron.

## ⚡ Fungsi Aktivasi
Hidden layer menggunakan **Sigmoid** (rumus: f(x) = 1 / (1 + e^-x)). Range output 0..1. Output layer tetap softmax.

## 2. Import library

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

from src.data_loader import load_iris_data
from src.utils import set_global_seed, plot_training_curves, save_history, evaluate_and_report
from src.config import EPOCHS, BATCH_SIZE, OPTIMIZER, LOSS, METRICS, VALIDATION_SPLIT, RANDOM_SEED

set_global_seed(RANDOM_SEED)
print('TensorFlow version:', tf.__version__)
```

## 3. Load data
Catatan: data sudah otomatis di-split, di-shuffle, dan dinormalisasi sesuai konfigurasi bersama di `src/config.py`. **Jangan diubah** supaya perbandingan adil.

```python
X_train, X_test, y_train, y_test, n_features, n_classes = load_iris_data()
print(f'Jumlah fitur: {n_features}, jumlah kelas: {n_classes}')
print(f'X_train: {X_train.shape}, X_test: {X_test.shape}')
```

## 4. Bangun model

```python
model = Sequential([
    Input(shape=(n_features,)),
    Dense(16, activation='sigmoid'),
    Dense(n_classes, activation='softmax'),
])
model.compile(optimizer=OPTIMIZER, loss=LOSS, metrics=METRICS)
model.summary()
```

## 5. Latih model
Hyperparameter (epochs, batch_size, optimizer) diambil dari `src/config.py` supaya identik dengan varian lain.

```python
history = model.fit(
    X_train, y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VALIDATION_SPLIT,
    verbose=2,
)
```

## 6. Visualisasi kurva training

```python
plot_training_curves(history, variant_name='02_mlp_sigmoid')
```

## 7. Evaluasi di test set

```python
summary = evaluate_and_report(model, X_test, y_test, variant_name='02_mlp_sigmoid')
save_history(history, variant_name='02_mlp_sigmoid')
summary
```

2. Apakah Sigmoid cocok untuk hidden layer? Coba kaitkan dengan fenomena 'vanishing gradient'.

Secara umum, Sigmoid kurang cocok digunakan pada hidden layer, terutama jika arsitektur jaringan semakin dalam (deep neural network). **Hal ini terjadi karena** Hal ini disebabkan oleh fenomena vanishing gradient. Fungsi aktivasi Sigmoid memetakan nilai input ke dalam rentang yang sangat sempit, yaitu antara 0 dan 1. Karakteristik ini membuat turunan (gradien) maksimum dari fungsi Sigmoid bernilai sangat kecil, yaitu hanya sebesar 0.25. Saat proses backpropagation berlangsung menggunakan chain rule, gradien dari layer output akan dikalikan dengan nilai turunan di setiap hidden layer. Perkalian berulang dari nilai yang kecil ini ($< 0.25$) menyebabkan nilai gradien mengecil secara eksponensial (lenyap) sebelum mencapai layer-layer awal. Akibatnya, bobot pada layer awal hampir tidak mengalami perubahan dan model berhenti belajar secara optimal. Untuk hidden layer, fungsi aktivasi seperti ReLU jauh lebih direkomendasikan.

3. Jika kamu menambah hidden layer kedua (lagi Sigmoid), apakah hasilnya membaik atau memburuk?

Hasilnya kemungkinan besar akan memburuk atau menjadi jauh lebih lambat untuk konvergen jika jumlah epoch dibatasi tetap 100. **Hal ini terjadi karena** Menambahkan hidden layer kedua yang kembali menggunakan Sigmoid akan memperparah efek vanishing gradient yang telah dijelaskan pada nomor 2. Gradien yang mengalir mundur ke hidden layer pertama harus melewati dua fungsi Sigmoid, yang berarti gradien akan tereduksi dua kali lebih cepat karena dikalikan berturut-turut dengan nilai turunan yang kecil. Akibatnya, pembaruan bobot (weight update) pada layer-layer awal menjadi sangat tidak signifikan, proses training melambat drastis, dan model akan membutuhkan jauh lebih banyak epoch hanya untuk mencapai tingkat akurasi yang sama dengan model 1 hidden layer saat ini.
